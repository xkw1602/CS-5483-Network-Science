```python
# Required Header
%matplotlib inline

import matplotlib.pyplot as plt
import networkx as nx
```

## Create a undirected graph UG and show it

```python
UG = nx.fast_gnp_random_graph(15,0.3)
nx.draw(UG,with_labels=True)
```

### The degree centrality values are normalized by dividing by the maximum possible degree in a simple graph n-1 where n is the number of nodes in UG.

```python
print("Determining the degree centrality of the above graph")
print("----------------------------------------------------")
d = nx.degree_centrality(UG)
for v in UG.nodes():
    print(f"{v:2} {d[v]:.3f}")
sort_list = sorted(d.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in UG:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(UG, node_color=color_map, with_labels=True)
```

### Katz Centrality

#### Katz centrality computes the centrality for a node based on the centrality of its neighbors. It is a generalization of the eigenvector centrality. The Katz centrality for node  is

##### =+, where  is the adjacency matrix of graph G with eigenvalues . The parameter  controls the initial centrality and  <1 max.
#### Katz centrality computes the relative influence of a node within a network by measuring the number of the immediate neighbors (first degree nodes) and also all other nodes in the network that connect to the node under consideration through these immediate neighbors. Extra weight can be provided to immediate neighbors through the parameter . Connections made with distant neighbors are, however, penalized by an attenuation factor  which should be strictly less than the inverse largest eigenvalue of the adjacency matrix in order for the Katz centrality to be computed correctly.

```python
print("Determining the Katz centrality of the above graph")
print("----------------------------------------------------")

centrality = nx.katz_centrality_numpy(UG, alpha=0.2, beta=1.0, normalized=True, weight=None)
for v, c in centrality.items():
    print(f"{v:2} {c:.3f}")

sort_list = sorted(centrality.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in UG:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(UG, node_color=color_map, with_labels=True)
```

### Closeness Centrality

#### Closeness centrality of a node u is the reciprocal of the average shortest path distance to u over all n-1 reachable nodes.

#### ()=1/(n-1) * (,), where d(v, u) is the shortest-path distance between v and u, and n is the number of nodes that can reach u. Notice that the closeness distance function computes the incoming distance to u for directed graphs. Notice that higher values of closeness indicate higher centrality.

```python
print("Determining the Closeness centrality of the above graph")
print("----------------------------------------------------")

centrality = nx.closeness_centrality(UG, u=None, distance=None, wf_improved=True)
for v, c in centrality.items():
    print(f"{v:2} {c:.3f}")

sort_list = sorted(centrality.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in UG:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(UG, node_color=color_map, with_labels=True)
```

## Create a directed graph DG and show it

```python
print("Determining the PageRank Centrality of the above Directed graph")
print("----------------------------------------------------")

centrality = nx.pagerank(DG, alpha=0.5, personalization=None, max_iter=100, tol=1e-06, nstart=None, weight='weight', dangling=None)
for v, c in centrality.items():
    print(f"{v:2} {c:.3f}")

sort_list = sorted(centrality.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in DG:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(DG, node_color=color_map, with_labels=True)
```

### The HITS algorithm computes two numbers for a node. Authorities estimates the node value based on the incoming links. Hubs estimates the node value based on outgoing links.

```python
h, a = nx.hits(DG, max_iter=100, tol=1e-08, nstart=None, normalized=True)

print("Determining the Hubs of the above Directed graph")
print("----------------------------------------------------")

for v, c in h.items():
    print(f"{v:2} {c:.3f}")
sort_list = sorted(h.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in DG:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(DG, node_color=color_map, with_labels=True)
```

```python
##------------------------

print("Determining the Authorities of the above Directed graph")
print("----------------------------------------------------")

for v, c in a.items():
    print(f"{v:2} {c:.3f}")
sort_list = sorted(a.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in DG:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(DG, node_color=color_map, with_labels=True)
```

### Acyclic Directed Graph

```python
DAG = nx.DiGraph()
DAG.add_edges_from([(1, 2), (1, 3), (2, 3), (1, 4), (4, 11), (4, 10), (3, 6), (6, 9), (6, 8), (2, 5), (5, 7)])
nx.draw_spectral(DAG, with_labels=True)
```

```python
print("Determining the Eigenvector centrality of the above graph")
print("----------------------------------------------------")

centrality = nx.eigenvector_centrality(DAG,max_iter=1000)
for v, c in centrality.items():
    print(f"{v:2} {c:.3f}")

sort_list = sorted(centrality.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in DAG:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(DAG, node_color=color_map, with_labels=True)
```

```python
print("Determining the Katz centrality of the above graph")
print("----------------------------------------------------")

centrality = nx.katz_centrality(DAG,alpha=0.3,beta=1.0)
for v, c in centrality.items():
    print(f"{v:2} {c:.3f}")

sort_list = sorted(centrality.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in DAG:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(DAG, node_color=color_map, with_labels=True)
```

### A k-core is a maximal subgraph that contains nodes of degree k or more

```python
NG = nx.fast_gnp_random_graph(10,0.5)

KG = nx.k_core(NG,k=4)

limits = plt.axis("off")  # turn of axis
plt.figure()
plt.subplot(121)
nx.draw(NG, with_labels=True)
plt.subplot(122)
nx.draw(KG, with_labels=True)
plt.show()
```
