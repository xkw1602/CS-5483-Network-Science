```python
G = nx.Graph()
G.add_edge('v1', 'v2', weight = 5)
G.add_edge('v1', 'v4', weight = 7)
G.add_edge('v2', 'v3', weight = 6)
G.add_edge('v2', 'v4', weight = 8)
G.add_edge('v2', 'v5', weight = 7)
G.add_edge('v3', 'v5', weight = 9)
G.add_edge('v4', 'v5', weight = 5)
G.add_edge('v4', 'v6', weight = 8)
G.add_edge('v5', 'v7', weight = 9)
G.add_edge('v5', 'v9', weight = 15)
G.add_edge('v6', 'v7', weight = 5)
G.add_edge('v6', 'v8', weight = 7)
G.add_edge('v7', 'v8', weight = 9)
G.add_edge('v7', 'v9', weight = 8)
G.add_edge('v8', 'v9', weight = 11)

pos = nx.spring_layout(G)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels)
```
