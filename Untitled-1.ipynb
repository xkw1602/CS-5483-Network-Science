```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.DiGraph()
G.add_node('Alice')
G.add_node('Bob')
G.add_node('Chuck')
list(G.nodes())
```

```python
G.add_edge('Alice', 'Bob')
G.add_edge('Alice', 'Chuck')
G.add_edge('Bob', 'Alice')
G.add_edge('Bob', 'Chuck')
list(G.edges())
```

```python
nx.draw_circular(G, node_color='C0', node_size=2000, with_labels=True)
plt.show()
```
